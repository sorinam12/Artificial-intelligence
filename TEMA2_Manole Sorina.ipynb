```python
import numpy as np 
import math 

x=[[45, 85],[50, 43],[40, 80],[55,42],[200, 43],[48, 40],[195,41],[43, 87],[190, 40]]

centroid1=[45,85]
centroid2=[50,43]
centroid3=[200, 43]
centroid1_NOU=[0,0]
centroid2_NOU=[0,0]
centroid3_NOU=[0,0]


cluster1=[]
cluster2=[]
cluster3=[]
j=1

while(1):
    print ("Runda ", j)
    j=j+1

    for i in range(len(x)):
        d1=abs(x[i][0]-centroid1[0])+abs(x[i][1]-centroid1[1])
        d2=abs(x[i][0]-centroid2[0])+abs(x[i][1]-centroid2[1])
        d3=abs(x[i][0]-centroid3[0])+abs(x[i][1]-centroid3[1])
        if d1<d2 and d1<d3:
            print ("Punctul[",x[i][0],",", x[i][1],"]"," apartine clusterului 1." )
            cluster1.append(x[i])

        elif d2<d1 and d2<d3:
            print ("Punctul[",x[i][0],",", x[i][1],"]"," apartine clusterului 2." )
            cluster2.append(x[i])
            #if d3<d2 and d3<d1:
        else:
            print ("Punctul[",x[i][0],",", x[i][1],"]"," apartine clusterului 3." )
            cluster3.append(x[i])

    
    print("cluster1=",cluster1)
    print("cluster2=",cluster2)
    print("cluster3=",cluster3)        
            
    sumaX1=0
    sumaX2=0
    sumaX3=0
    sumaY1=0
    sumaY2=0
    sumaY3=0

    for i in range(0,len(cluster1)):
        sumaX1=sumaX1+cluster1[i][0]
        sumaY1=sumaY1+cluster1[i][1]
    for i in range(0,len(cluster2)):
        sumaX2=sumaX2+cluster2[i][0]
        sumaY2=sumaY2+cluster2[i][1]
    for i in range(0,len(cluster3)):
        sumaX3=sumaX3+cluster3[i][0] 
        sumaY3=sumaY3+cluster3[i][1]


    MedXcentroid1=sumaX1/len(cluster1)
    MedYcentroid1=sumaY1/len(cluster1)
    MedXcentroid2=sumaX2/len(cluster2)
    MedYcentroid2=sumaY2/len(cluster2)
    MedXcentroid3=sumaX3/len(cluster3)
    MedYcentroid3=sumaY3/len(cluster3)
    
    cluster1.clear()
    cluster2.clear()
    cluster3.clear()
    
    print("centroid1=",centroid1)
    print("centroid2=",centroid2)
    print("centroid3=",centroid3)
    
    centroid1_NOU[0]=MedXcentroid1
    centroid1_NOU[1]=MedYcentroid1
    centroid2_NOU[0]=MedXcentroid2
    centroid2_NOU[1]=MedYcentroid2
    centroid3_NOU[0]=MedXcentroid3
    centroid3_NOU[1]=MedYcentroid3

    print("centroid1 nou=",centroid1_NOU)
    print("centroid2 nou=",centroid2_NOU)
    print("centroid3 nou=",centroid3_NOU)


    if centroid1_NOU==centroid1 and centroid2_NOU==centroid2 and centroid3_NOU==centroid3: break

    centroid1=centroid1_NOU
    centroid2=centroid2_NOU
    centroid3=centroid3_NOU
    
```
