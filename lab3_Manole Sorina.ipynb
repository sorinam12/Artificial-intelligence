```python
x = 3 
c = 0.01 
precizie = 0.000001 
unitate = 1 
pas_max = 10000
pas = 0 
df = lambda x: 12*x-12

while unitate > precizie and pas < pas_max:
    prev_x = x 
    x = x - c * df(prev_x) 
    unitate = abs(x - prev_x) 
    pas = pas+1 
    print("Iteratia",pas,"\nX este:",x) 
    
print("Minimul local este: ", x)
```

```python
x = 3 
y = 3
c = 0.01 
precizie = 0.000001 
unitate_x = 1
unitate_y = 1
pas_max = 15000
pas = 0 
dfx = lambda x: 2*x
dfy = lambda y: 4*y

while unitate_x > precizie and unitate_y > precizie and pas < pas_max:
    prev_x = x 
    prev_y = y 
    
    x = x - c * dfx(prev_x) 
    y = y - c * dfy(prev_y)
    unitate_x = abs(x - prev_x)
    unitate_y = abs(y - prev_y)
    
    pas = pas+1 
    print("Iteratia",pas,"\nX este:",x,"\nY este:",y ) 
    
print("Minimul local pentru X este: ", x)
print("\nMinimul local pentru Y este: ", y)
```
